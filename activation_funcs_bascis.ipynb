```python
# import libraries
import torch
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size':18})
```

```python
x = torch.linspace(-3,3,101)

def NNoutput(actfun):
  actfun= getattr(torch,actfun)
  return actfun(x)
```

```python
act_funs=['relu','tanh','sigmoid']
for act in act_funs:
  y=NNoutput(act)
  plt.plot(x,y)
  plt.legend(['relu','tanh','sigmoid'])
```
